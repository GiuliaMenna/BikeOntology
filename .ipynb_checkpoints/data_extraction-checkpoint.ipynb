```python
import pandas as pd

df = pd.read_csv('data 99Spokes/Sample Bike Specs - 2022-07-22.csv', sep=";")
```

```python
df = df.where((pd.notnull(df)), None)
df
```

```python
print(df.columns)
```

```python
nameofPropandInverse={"type":{"name":"hasType", "inverse":"isTypeOf"},
                     "brand":{"name":"hasBrand", "inverse":"isBrandOf"}}
```

```python
dict_ontology = {"Maker":["brand", "Brand"], "Year":"vso:modelDate", "Category":["type", "TypeOfBike"], 
                 "Frame [Raw]":"Frame", "Fork [Raw]":"Fork", "RD [Raw]":"RearDerailleur", 
                 "FD [Raw]":"FrontDerailleur", "Crank [Raw]":"Crank", "Cassette [Raw]":"Cogset", 
                 "Chain [Raw]":"Chain", "Brakes [Raw]":"Brakes", "Wheelset [Raw]":"Wheel", 
                 "Tires [Raw]":"Tire", "Handlebar [Raw]":"HandlebarGrip", "Seatpost [Raw]":"SeatPost", 
                 "Saddle [Raw]":"Saddle", "Weight [KG]":"vso:weight", "Motor [Raw]":"Motor", "Battery [Raw]":"Battery"}
```

```python
from rdflib import URIRef, BNode, Literal, Graph
from rdflib.namespace import CSVW, DC, DCAT, DCTERMS, DOAP, FOAF, ODRL2, ORG, OWL, \
                           PROF, PROV, RDF, RDFS, SDO, SH, SKOS, SOSA, SSN, TIME, \
                           VOID, XMLNS, XSD
from rdflib import Namespace
```

```python
bikeo = Namespace("http://purl.org/ontology/bikeo#")   #we use this to declare our prefix in the graph
vso = Namespace("http://purl.org/vso/ns#") #prefix for the vehicle sales ontology
g = Graph() #we create the graph
g.bind("bikeo", bikeo)
g.bind("rdf",  RDF)
g.bind("rdfs", RDFS)
g.bind("vso", vso)

for idx, bike in df.iterrows(): #iterating over the whole dataframe
    g.add((URIRef("http://purl.org/ontology/bikeo#"+bike['ID']), RDF.type, bikeo.Bike)) #add each bike to the graph
    g.add((URIRef("http://purl.org/ontology/bikeo#"+bike['ID']), RDFS.label, Literal(bike["Maker"]+" "+bike["Model"]+" "+str(bike["Year"]))))
    
    if bike["Gearing [Configuration]"] is not None:
        g.add((URIRef("http://purl.org/ontology/bikeo#"+bike['ID']), bikeo.hasGearing, Literal(bike["Gearing [Configuration]"])))
    for el in dict_ontology:
        if bike[el]: #checking if there is a column with the name in the dict in the dataframe
            if type(dict_ontology[el]) is list: #handling the case for the properties that have an inverse 
                name = nameofPropandInverse[dict_ontology[el][0]]["name"] # we take the name of the relationship from the dictionary we have created
                inverse = nameofPropandInverse[dict_ontology[el][0]]["inverse"] #as well as the inverse

                if el == "Category" and bike["E-Bike"] == "VERO": #for electric bikes, use the column "E-Bike"
                    #adding the triples with both the property and its inverse
                    g.add((URIRef("http://purl.org/ontology/bikeo#"+bike['ID']), URIRef("http://purl.org/ontology/bikeo#"+name), URIRef("http://purl.org/ontology/bikeo#"+"electric"+bike[el])))
                    g.add((URIRef("http://purl.org/ontology/bikeo#"+"electric"+bike[el]), URIRef("http://purl.org/ontology/bikeo#"+inverse), URIRef("http://purl.org/ontology/bikeo#"+bike['ID'])))
                    g.add((URIRef("http://purl.org/ontology/bikeo#"+"electric"+bike[el]), RDF.type, URIRef("http://purl.org/ontology/bikeo#"+dict_ontology[el][1])))
                    g.add((URIRef("http://purl.org/ontology/bikeo#"+"electric"+bike[el]), RDFS.label, Literal("Electric "+bike[el])))
                else:
                    var = bike[el].lower().strip()
                    #adding the triples with both the property and its inverse
                    g.add((URIRef("http://purl.org/ontology/bikeo#"+bike['ID']), URIRef("http://purl.org/ontology/bikeo#"+name), URIRef("http://purl.org/ontology/bikeo#"+var)))
                    g.add((URIRef("http://purl.org/ontology/bikeo#"+var), URIRef("http://purl.org/ontology/bikeo#"+inverse), URIRef("http://purl.org/ontology/bikeo#"+bike['ID'])))
                    #adding the brand or type of the specific bike as the type in the ontology
                    g.add((URIRef("http://purl.org/ontology/bikeo#"+var), RDF.type, URIRef("http://purl.org/ontology/bikeo#"+dict_ontology[el][1])))
                    g.add((URIRef("http://purl.org/ontology/bikeo#"+var), RDFS.label, Literal(bike[el])))
            elif ":" in dict_ontology[el]: #special case if the property comes from vso (the only properties with : inside the dictionary) 
                prop = dict_ontology[el].replace("vso:", "")
                if type(bike[el]) is str:
                    val = bike[el].replace(",",".")
                else:
                    val = bike[el]
                if prop == "modelDate":
                    g.add((URIRef("http://purl.org/ontology/bikeo#"+bike['ID']), URIRef("http://purl.org/vso/ns#"+prop), Literal(val,datatype=XSD.integer)))
                else:
                    g.add((URIRef("http://purl.org/ontology/bikeo#"+bike['ID']), URIRef("http://purl.org/vso/ns#"+prop), Literal(val,datatype=XSD.float)))
            else:
                #cleaning some content
                var = bike[el].replace("|", ",")
                var = var.replace("(",",").replace(")", ",")
                #some fields with [Raw] contain lots of info separated by commas
                parts = var.split(",")
                #selecting only the first, usually the name of the part
                part = parts[0]
                
                #verifying that the first element is what we are looking for, cleaning it and selecting the right one otherwise
                if part.startswith("Size:"):
                    for elmt in parts[1:]:
                        if elmt.strip().isdigit() is False:
                            part = elmt
                            break
                if part.startswith("Front: "):
                    part = part.strip("Front: ")
                #some parts had description of sizes. Removing it
                if part.endswith("S:40cm"):
                    if part.endswith(" XS:40cm"):
                        part = part.strip(" XS:40cm")
                    else:
                        part = part.strip(" S:40cm")
                #keeping the string stored to add it as literal of the subject
                label = part.strip()
                #transform it to add it to the ontology
                part = part.lower().strip().replace(" ", "-").replace(":","-").replace("/", "-").replace("™","").replace("'", "-")
                part += "-"+dict_ontology[el].lower()
                #add the part as part of the bike (property and the inverse), register it of the general type and add the label.
                g.add((URIRef("http://purl.org/ontology/bikeo#"+bike['ID']), bikeo.hasPart, URIRef("http://purl.org/ontology/bikeo#"+part)))
                g.add((URIRef("http://purl.org/ontology/bikeo#"+part), bikeo.isPartOf, URIRef("http://purl.org/ontology/bikeo#"+bike['ID'])))
                g.add((URIRef("http://purl.org/ontology/bikeo#"+part), RDF.type, URIRef("http://purl.org/ontology/bikeo#"+dict_ontology[el])))
                g.add((URIRef("http://purl.org/ontology/bikeo#"+part), RDFS.label, Literal(label)))
                
                part_brand = el.split()[0] + " [Maker]"
                if part_brand in df.columns and bike[part_brand] is not None:
                    bike_part_brand = bike[part_brand].lower().strip().replace(" ", "-").replace(":","-").replace("/", "-").replace("™","").replace("'", "-")
                    g.add((URIRef("http://purl.org/ontology/bikeo#"+part), bikeo.producedBy, URIRef("http://purl.org/ontology/bikeo#"+bike_part_brand)))
                    g.add((URIRef("http://purl.org/ontology/bikeo#"+bike_part_brand), bikeo.produces, URIRef("http://purl.org/ontology/bikeo#"+part)))
                    g.add((URIRef("http://purl.org/ontology/bikeo#"+bike_part_brand), RDF.type, bikeo.Brand))
                    g.add((URIRef("http://purl.org/ontology/bikeo#"+bike_part_brand), RDFS.label, Literal(bike[part_brand].strip())))
                    
                #adding tire size and chainring to crank which are data properties
                if el == "Tires [Raw]" and bike["Tires [Width]"] is not None:
                    g.add((URIRef("http://purl.org/ontology/bikeo#"+part), bikeo.hasSize, Literal(bike["Tires [Width]"])))
                
                elif el == "Crank [Raw]" and bike["Chainring [Gearing]"] is not None:
                    chainring = bike["Chainring [Gearing]"]
                    if "–" in chainring:
                        chainring = chainring.split("–")
                        for el in chainring:
                            g.add((URIRef("http://purl.org/ontology/bikeo#"+part), bikeo.hasChainRing, Literal(el,datatype=XSD.integer)))
                    else:
                        g.add((URIRef("http://purl.org/ontology/bikeo#"+part), bikeo.hasChainRing, Literal(chainring,datatype=XSD.integer)))
```

```python
print(g.serialize(format="turtle")) #this prints the serialization in turtle in the console
# we can save the final turtle like this
#uncomment the last line to save the file in the repository
g.serialize(destination = "abox_bikeo.ttl", format="turtle") # this creates the final turtle
```
